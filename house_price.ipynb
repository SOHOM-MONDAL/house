```python
import pandas as pd
import matplotlib.pyplot as  plt
df = pd.read_csv('Downloads/house price.csv')
df.head()
```

```python
df.pop("waterfront")
#plt.scatter(X,y)
df.pop("date")
df.pop("view")
df.pop("condition")
df.pop("yr_renovated")
df.pop("yr_built")
df.pop("city")
df.pop("statezip")
df.pop("country")
df.pop("street")
df.head()
print(df.describe())
```

```python
fatures_names =['bedrooms'	,'bathrooms'	,'sqft_living'	, 'sqft_lot' ,'floors', 'sqft_above','sqft_basement']
target_names = 'price'
X =df[fatures_names]
y =df[target_names]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
import numpy as np
mod = reg.fit(X_train, y_train)
pred = reg.predict(X_test)
pred
```

```python
plt.scatter(y_test,pred,marker='*')
plt.ylabel('price predicted',fontsize=50)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
score = r2_score(y_test,pred)
print("Test R2 value",score)
```

```python
mse = mean_squared_error(pred,y_test)
print("mean square error value",mse)
```
